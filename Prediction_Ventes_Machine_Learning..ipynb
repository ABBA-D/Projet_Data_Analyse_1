```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("historique_ventes.csv")
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data['Date_Num'] = data['Date'].map(pd.Timestamp.toordinal)
X = data[['Date_Num']]
y = data['Montant_Total']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
print("Erreur quadratique moyenne (MSE) :", mse)
```

```python
plt.scatter(X_test, y_test, color='blue', label="Réel")
plt.plot(X_test, y_pred, color='red', label="Prédiction")
plt.xlabel("Date")
plt.ylabel("Montant Total des Ventes")
plt.title("Prédictions des Ventes avec Régression Linéaire")
plt.legend()
plt.show()
```
