```python
import pandas as pd

# Charger le fichier CSV des ventes de produits
data_ventes = pd.read_csv("ventes_produits.csv")
data_ventes.head()
```

```python
data_ventes.info()
data_ventes.isnull().sum()  # Compter les valeurs manquantes par colonne
```

```python
data_ventes.describe()
```

```python
data_ventes['Produit'].value_counts().plot(kind='bar', title="Nombre de ventes par produit")
```

```python
data_ventes.groupby('Produit')['Montant_Total'].sum().plot(kind='pie', autopct='%1.1f%%', title="Répartition du montant total des ventes par produit")
```

```python
data_ventes.plot(kind='scatter', x='Quantité', y='Montant_Total', title="Relation entre Quantité et Montant Total")
```
